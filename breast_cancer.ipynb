importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

data collection and processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the  data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
#adding the "target"column to the frame
data_frame["label"]= breast_cancer_dataset.target
```

separating the features and target

```python
x= data_frame.drop(columns="label", axis=1)
y= data_frame["label"]
```

```python
print(x)
```

```python
print(y)
```

splitting the data  into training & testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

standardize the data


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

building the neural network

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation="relu"),
    keras.layers.Dense(2, activation="sigmoid")
])
```

```python
#compelling the neural network

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# splitting the training data into training and validation data
X_train_split, X_val, Y_train_split, Y_val = train_test_split(X_train_std, Y_train, test_size=0.1, random_state=2)

# training  the neural network
history = model.fit(X_train_split, Y_train_split, validation_data=(X_val, Y_val), epochs=10)
```

visualizing accuracy and  loss

```python
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["training data","validation"],loc = "lower right")
plt.show()
```

```python
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.title("model loss")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.legend(["training data","validation"],loc = "upper right")
plt.show()
```

accuracy of the on test data

```python
loss , accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
